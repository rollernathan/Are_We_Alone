```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

# Step 1: Data Extraction
url = 'https://nuforc.org/ndx/?id=event'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing the data
table = soup.find('table')

# Extract data from the table
data = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    cols = row.find_all('td')
    if len(cols) == 2:
        date = cols[0].text.strip()
        count = cols[1].text.strip()
        data.append([date, count])

# Step 2: Data Preparation
df = pd.DataFrame(data, columns=['Date', 'Count'])

# Print to identify any problematic date formats
print("Unique Date Formats:", df['Date'].unique())

# Convert Date to datetime with errors='coerce' to handle out-of-bound dates
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m', errors='coerce')
df['Count'] = pd.to_numeric(df['Count'], errors='coerce')

# Drop rows where Date is NaT (invalid dates)
df = df.dropna(subset=['Date'])

# Step 3: Data Visualization
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Count'], marker='o', linestyle='-')
plt.title('Monthly UFO Sightings Reported to NUFORC (2000-2024)')
plt.xlabel('Date')
plt.ylabel('Number of Sightings')
plt.grid(True)
plt.show()
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

# Step 1: Data Extraction
url = 'https://nuforc.org/ndx/?id=event'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing the data
table = soup.find('table')

# Extract data from the table
data = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    cols = row.find_all('td')
    if len(cols) == 2:
        date = cols[0].text.strip()
        count = cols[1].text.strip()
        data.append([date, count])

# Step 2: Data Preparation
df = pd.DataFrame(data, columns=['Date', 'Count'])
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m', errors='coerce')
df['Count'] = pd.to_numeric(df['Count'], errors='coerce')

# Drop rows with invalid dates
df = df.dropna(subset=['Date'])

# Extract the year and determine the decade
df['Year'] = df['Date'].dt.year
df['Decade'] = (df['Year'] // 10) * 10

# Calculate average monthly sightings per decade
decade_avg_counts = df.groupby('Decade')['Count'].mean()

# Step 3: Bar Chart Visualization
plt.figure(figsize=(10, 6))
decade_avg_counts.plot(kind='bar')
plt.title('Average Monthly UFO Sightings per Decade')
plt.xlabel('Decade')
plt.ylabel('Average Monthly Sightings')
plt.xticks(rotation=45)
plt.show()
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

# Step 1: Data Extraction
url = 'https://nuforc.org/ndx/?id=event'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing the data
table = soup.find('table')

# Extract data from the table
data = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    cols = row.find_all('td')
    if len(cols) == 2:
        date = cols[0].text.strip()
        count = cols[1].text.strip()
        data.append([date, count])

# Step 2: Data Preparation
df = pd.DataFrame(data, columns=['Date', 'Count'])
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m', errors='coerce')
df['Count'] = pd.to_numeric(df['Count'], errors='coerce')

# Drop rows with invalid dates
df = df.dropna(subset=['Date'])

# Filter data for the years 2020 to 2024
df_filtered = df[(df['Date'].dt.year >= 2020) & (df['Date'].dt.year <= 2024)]

# Step 3: Bar Chart Visualization
plt.figure(figsize=(14, 7))
plt.bar(df_filtered['Date'], df_filtered['Count'], width=20, align='center')
plt.title('Monthly UFO Sightings from 2020 to 2024')
plt.xlabel('Date')
plt.ylabel('Number of Sightings')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

# Step 1: Data Extraction
url = 'https://nuforc.org/ndx/?id=event'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing the data
table = soup.find('table')

# Extract data from the table
data = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    cols = row.find_all('td')
    if len(cols) == 2:
        date = cols[0].text.strip()
        count = cols[1].text.strip()
        data.append([date, count])

# Step 2: Data Preparation
df = pd.DataFrame(data, columns=['Date', 'Count'])
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m', errors='coerce')
df['Count'] = pd.to_numeric(df['Count'], errors='coerce')

# Drop rows with invalid dates
df = df.dropna(subset=['Date'])

# Filter data for the year 2024 and group by month
df_2024 = df[df['Date'].dt.year == 2024]
monthly_counts_2024 = df_2024.groupby(df_2024['Date'].dt.month)['Count'].sum()

# Step 3: Bar Chart Visualization
plt.figure(figsize=(10, 6))
monthly_counts_2024.plot(kind='bar')
plt.title('Monthly UFO Sightings in 2024')
plt.xlabel('Month')
plt.ylabel('Number of Sightings')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()
```
